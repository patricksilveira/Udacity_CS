```python
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the compareTriplets function below.
def compareTriplets(a, b):
    score_bob = 0
    score_ana = 0
    check = 0
    while check < len(a):
        if a[check] > b[check]:
            score_bob += +1 
            print (score_bob, score_ana)
        if a[check] < b[check]:
            score_ana += +1 
            print (score_bob, score_ana)
        check += +1
    return score_bob, score_ana
        
        

a = (17, 28, 30)
b = (99, 16, 8)

compareTriplets(a, b)
```

```python
import math
import os
import random
import re
import sys

def diagonalDifference(arr):
    diag1 = 0
    diag2 = 0
    count = 0
    reverse_count = -1
    while count < len(arr):
        diag1 = diag1 + arr[count][count]
        diag2 = diag2 + arr[count][reverse_count]
        count += +1
        reverse_count += -1
    return abs (diag2 - diag1)
    
    
    
#    new_list = ()
#   return (arr[0][0], arr[1][1], arr [2][2])

    
    
    
    
arr = [[11, 2, 4], [4, 5, 6], [10, 8, -12]]
print (diagonalDifference(arr))
```
